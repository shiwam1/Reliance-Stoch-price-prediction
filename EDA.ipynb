## 2. Importing Dataset

```python
data = pd.read_csv("D:/Project-3/Reliance_stock_price.csv")
data
```

```python
data = data.drop("Unnamed: 0", axis=1)
data
```

Observations :-

The dataset used is Reliance Industries stock market data from NSE (National Stock Exchange) India over the last 7 years (2015 - 2022)

The dataset contains 1976 samples. Each sample contains 7 features.

1. Date - Date of trade


2. Open - The open is the starting period of trading on a securities exchange or organized over-the-counter market.


3. High - Highest price at which a stock traded during the course of the trading day.


4. Low - Lowest price at which a stock traded during the course of the trading day.


5. Close - (Target variable) The close is a reference to the end of a trading session in the financial markets when the markets close for the day.


6. Adj.close - The price of the stock after paying off the dividends.


7. Volume - It is the amount of a security that was traded during a given period of time

## 4. Visualization

### 4.1 Histogram for 'Open', 'High', 'Low' , Close'

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,18))

plt.subplot(221)
data["Open"].hist(color='navy')
plt.xlabel("Opning Trend")
plt.ylabel("frequency")
plt.title ("Histogram for Opning Trend of Stock", color='brown')

plt.subplot(222)
data["High"].hist(color='navy')
plt.xlabel("Highest Price")
plt.ylabel("frequency")
plt.title ("Histogram for Highest Price of Stock", color='brown')

plt.subplot(223)
data["Low"].hist(color='navy')
plt.xlabel("Lowest Price")
plt.ylabel("frequency")
plt.title ("Histogram for Lowest Pric of Stock", color='brown')

plt.subplot(224)
data["Close"].hist(color='navy')
plt.xlabel("Closing Price")
plt.ylabel("frequency")
plt.title ("Histogram for Closing Price of Stock", color='brown')

plt.show()
```

Observations :-

1. By analysing the above histograms we can say that the given dataset is not symmetric in nature.

### 4.2 Boxplot for 'Open', 'High', 'Low' , Close'

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

numerical_features_name = data[["Open", "High", "Low", "Close"]]

i=0
fig = make_subplots( rows=1, cols = numerical_features_name.shape[1],subplot_titles = numerical_features_name.columns.values )
for feature in numerical_features_name.columns.values:
    fig.add_trace(go.Box( y=numerical_features_name[feature]), row=1,col=i+1 )
    i = i+1

fig.update_layout( width=1000, height=500)
fig.show()
```

Observations :-

1. After analysing all the above box plots we can see that the given dataset does not contains noisy data points or outliers.

### 4.3 'Open', 'High', 'Low' , Close' over the Time

```python
cols_plot = ['Open', 'High','Low', 'Close']
axes = data[cols_plot].plot(figsize=(15, 12), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily trade')
    ax.set_xlabel('Date')
```

Observations :-

1. After analyzing all the above graphs we can say that all the trend follows the same pattern over the last 7 years.

### 4.4 Kernel Distribution Estimation Plot (KDE)

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,18))

plt.subplot(221)
sns.kdeplot(data['Open'], shade=True, color='navy')

plt.subplot(222)
sns.kdeplot(data['High'], shade=True, color='navy')

plt.subplot(223)
sns.kdeplot(data['Low'], shade=True, color='navy')

plt.subplot(224)
sns.kdeplot(data['Close'], shade=True, color='navy')

plt.show()
```

Observations :-

1. For all the features density curve is right skewed and all the trend follows the same pattern.

```python
fig = go.Figure([go.Scatter(x=data['Date'], y=data['Close'])])
fig.update_layout(
    autosize=False,
    width=1000,
    height=500,
    title='Closing Price over time',
    template="simple_white",
)

fig.update_xaxes(title="Date")
fig.update_yaxes(title="Close")
fig.show()
```
